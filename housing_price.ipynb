```python
import matplotlib.pyplot as plt
import seaborn as sns

# Original dataset (with capped values at 5.0)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# ---- Left: With capped values ----
sns.regplot(x="MedInc", y="price", data=dataset,
            scatter_kws={"s": 10, "alpha": 0.5},
            line_kws={"color": "red"},
            ax=axes[0])
axes[0].set_title("With capped values (price=5.0)")

# ---- Right: After removing capped values ----
dataset_uncapped = dataset[dataset["price"] < 5.0]

sns.regplot(x="Population", y="price", data=dataset_uncapped,
            scatter_kws={"s": 10, "alpha": 0.5},
            line_kws={"color": "red"},
            ax=axes[1])
axes[1].set_title("Without capped values")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Split figure into 3 plots
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# ---- 1. Linear Regression (with capped values) ----
sns.regplot(x="MedInc", y="price", data=dataset,
            scatter_kws={"s": 10, "alpha": 0.5},
            line_kws={"color": "red"},
            ax=axes[0])
axes[0].set_title("Linear (with capped values)")

# ---- 2. Linear Regression (without capped values) ----
dataset_uncapped = dataset[dataset["price"] < 5.0]

sns.regplot(x="MedInc", y="price", data=dataset_uncapped,
            scatter_kws={"s": 10, "alpha": 0.5},
            line_kws={"color": "red"},
            ax=axes[1])
axes[1].set_title("Linear (without capped values)")

# ---- 3. Polynomial Regression (order=2) ----
sns.regplot(x="MedInc", y="price", data=dataset_uncapped,
            order=2,  # quadratic fit
            scatter_kws={"s": 10, "alpha": 0.5},
            line_kws={"color": "green"},
            ax=axes[2])
axes[2].set_title("Polynomial Regression (order=2)")

plt.show()
```

```python
#independent and non independent features
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,]
```

```python
x.head()
```

```python
y.head()
```

# we try train test split for Regression model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_test
```

```python
x_train
```

# Standardrize our test and train sets

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
#fit_transform is not used with test data for our model does'nt have much information about standardized train_set, our aim is to get to the global minima
x_test=scaler.transform(x_test)
```

# model training will be done through linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
print("Intercepts:", np.round(regression.intercept_, 3))
```

```python
##parameterss the model is trained
regression.get_params()
```

```python
#prediction with test data
reg_pred=regression.predict(x_test)
```

# Assumptions

```python
reg_pred
```

```python
##scater plot for prediction
plt.scatter(reg_pred,y_test)
```

```python
residuals=y_test-reg_pred
```

```python
residuals
```

```python
##plotting the residuals
sns.kdeplot(residuals, shade=True)
plt.show()
```

```python
##scatterplot wrt predictions and residuals
plt.scatter(reg_pred,residuals)
```

```python
plt.scatter(reg_pred, residuals, alpha=0.6)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlim(0, 2000)   # adjust based on your data
plt.show()
```

```python
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(8,5))
plt.scatter(reg_pred, residuals, alpha=0.5, edgecolors="k")
plt.axhline(y=0, color="red", linestyle="--")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted Values")
plt.show()
```

# Performance matrices

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
print(r2_score(y_test, reg_pred))
```
